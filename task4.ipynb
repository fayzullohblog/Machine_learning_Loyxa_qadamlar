```python
import pandas as pd

boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'

boston_df = pd.read_csv(boston_url)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Boxplot for "Median value of owner-occupied homes":
plt.figure(figsize=(8, 6))
plt.boxplot(boston_df['MEDV'])
plt.xlabel('Homes')
plt.ylabel('Median Value ($1000s)')
plt.title('Boxplot of Median Value of Owner-Occupied Homes')
plt.show()
```

```python
# Bar plot for the "Charles river" variable:
charles_river_counts = boston_df['CHAS'].value_counts()
plt.bar(charles_river_counts.index, charles_river_counts.values)
plt.xlabel('Charles River')
plt.ylabel('Frequency')
plt.title('Bar Plot of Charles River Variable')
plt.xticks([0, 1], ['Not Bound', 'Bound'])
plt.show()
```

```python
# Boxplot for "Median value of owner-occupied homes" vs "AGE" variable:
age_groups = pd.cut(boston_df['AGE'], bins=[0, 35, 70, np.inf], labels=['35 years and younger', 'between 35 and 70 years', '70 years and older'])
plt.figure(figsize=(8, 6))
sns.boxplot(x=age_groups, y=boston_df['MEDV'])
plt.xlabel('Age Groups')
plt.ylabel('Median Value ($1000s)')
plt.title('Boxplot of Median Value of Owner-Occupied Homes by Age Group')
plt.show()
```

```python
# Scatter plot for "Nitric oxide concentrations" vs "Proportion of non-retail business acres per town":
plt.figure(figsize=(8, 6))
plt.scatter(boston_df['NOX'], boston_df['INDUS'])
plt.xlabel('Nitric Oxide Concentrations (NOX)')
plt.ylabel('Proportion of Non-Retail Business Acres (INDUS)')
plt.title('Scatter Plot: NOX vs INDUS')
plt.show()
```

```python
# Histogram for the "Pupil-Teacher Ratio" variable:
plt.figure(figsize=(8, 6))
plt.hist(boston_df['PTRATIO'], bins=10, edgecolor='black')
plt.xlabel('Pupil-Teacher Ratio (PTRATIO)')
plt.ylabel('Frequency')
plt.title('Histogram of Pupil-Teacher Ratio')
plt.show()
```
