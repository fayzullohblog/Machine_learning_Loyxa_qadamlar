```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
df['median_income'].hist(bins=50,figsize=(9,6))
plt.show()
```

```python
df['income_cut']=pd.cut(df['median_income'],bins=[0.,1.5,3.0,4.5,5.0,np.inf],labels=[1,2,3,4,5])
df['income_cut'].hist(figsize=(9,6))
plt.show()
```

```python
df
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
stratified_split=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=40)
for train_index,test_index in stratified_split.split(df,df['income_cut']):
  strat_train_set=df.loc[train_index]
  strat_test_set=df.loc[test_index]
```

```python
strat_train_set['income_cut'].hist(figsize=(9,6))
plt.show()
```

```python
strat_train_set.drop('income_cut',axis=1,inplace=True)
strat_test_set.drop('income_cut',axis=1,inplace=True)
```

```python
housing=strat_train_set.copy()
housing
```

```python
housing.plot(kind='scatter',x='longitude',y='latitude')
plt.show()
```

```python
housing.plot(
    kind='scatter',
    x='longitude',
    y='latitude',
    alpha=0.4,
    s=housing['population']/50,
    c='median_house_value',
    cmap='jet',
    colorbar=True,
    figsize=(10,8)

    )
plt.show()
```

```python
housing.corrwith(housing['median_house_value']).sort_values(ascending=False)
```
